```python

# # Importing the adfuller function from the statsmodels library, which is used to perform the Augmented Dickey-Fuller (ADF) test.
# Function to print out results in customised manner
from statsmodels.tsa.stattools import adfuller

# Define a function named 'adf_test' that takes three parameters: timeseries, df, and ohlc_type.
# ADF Test: The function performs an Augmented Dickey-Fuller test to check for stationarity in a time series.
def adf_test(timeseries,df,ohlc_type):

    # Set the figure size for the plot to be created later.
    plt.figure(figsize=(16,5))
    # Plot the data
    # ADF Test
    # Function to print out results in customised manner
    from statsmodels.tsa.stattools import adfuller
    print ('Results of Dickey-Fuller Test:')

    # Perform the ADF test on the given time series data, using 'AIC' for automatic lag selection.
    dftest = adfuller(timeseries, autolag='AIC')
    
    # Print the raw results from the ADF test, which includes several values such as test statistic, p-value, etc.
    print(dftest)

    # Create a pandas Series to store the first four elements of the ADF test result.
    # These elements are: test statistic, p-value, number of lags used, and the number of observations.
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])

    # Add the critical values (1%, 5%, and 10%) to the dfoutput series by iterating over the dictionary in dftest[4].
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value

    # Print the formatted results stored in dfoutput.
    print (dfoutput)

    # Store the result series in a variable named 'ans'.
    ans=dfoutput

    # Print out the conditions to interpret the results of the test.
    print("Condition:")
    print("p-value<=0.05-->Accept Alternate Hypothesis")
    print("p-value>0.05-->Accept Null Hypothesis")

    # Check if the test statistic is less than any of the critical values and the p-value is less than 0.05. 
    # If the condition true , then it is a Stationarity based on ADH orelse it is an Non-stationarity based on ADH
    if(ans['Test Statistic']<ans["Critical Value (1%)"] or
       ans['Test Statistic']<ans["Critical Value (5%)"] or 
       ans['Test Statistic']<ans["Critical Value (10%)"]
       and ans["p-value"]<0.05):
        print("Condition: statictic < any critical value and p-value <0.05 to reject null hypothsis")
        print("Reject null hypothesis:Non Stationarity")
        print("Accept Alternate hypothesis:Staionarity ")
        message="Stationarity based on ADH"
    else:
        print("Condition: statictic < any critical value and p-value <0.05 to reject null hypothsis")
        print("Accept null hypothesis:Non Stationarity" )
        print("Reject Alternate hypothesis:Staionarity ")
        message="Non-stationarity based on ADH"

    # Plot the given data (df) with the index on the x-axis and the ohlc_type on the y-axis.
    plt.plot(df.index, df[ohlc_type], label = ohlc_type)

    # Add a legend to the plot, positioning it in the best location.
    plt.legend(loc='best')

    # Set the title of the plot using the message (stationarity or non-stationarity) and the pollutant name.
    plt.title("{}_{}_2013 to 2021".format(message,ohlc_type))

    # Save the plot as a PNG file with the name format "{ohlc_type}_ADF.png".
    plt.savefig("{}_ADH.png".format(ohlc_type))

    # Display the plot.
    plt.show()

    # Return the message indicating whether the series is stationary or non-stationary.
    return message
```

```python
adf_test(dataset["Close"],dataset,"Close")
```

```python
# ApEn is a function that calculates the Approximate Entropy (ApEn) of a time series U.
# Measures the regularity and unpredictability of fluctuations in a time series; a lower value indicates more regularity
# while a higher value indicates more randomness
def ApEn(U, m, r):
    """Compute Aproximate entropy
    U: The input time series (a list or array of data points).
    m: The embedding dimension, which defines the length of sequences to compare in the time series.
    r: The tolerance level, which defines the threshold for similarity between sequences."""
    def _maxdist(x_i, x_j):
        """_maxdist: This inner function computes the maximum distance between two vectors x_i and x_j.
        x_i: The first vector (a sequence of length m).
        x_j: The second vector (another sequence of length m).
        Returns: The maximum absolute difference between corresponding elements of x_i and x_j.
        """
        # Pairs corresponding elements of x_i and x_j using zip Computes the absolute difference between elements
        # Finds the maximum of these absolute differences.
        return max([abs(ua - va) for ua, va in zip(x_i, x_j)])

    def _phi(m):
        """
        This inner function computes a value used in the calculation of Approximate Entropy.
        m: The embedding dimension.
        x: A list of vectors, where each vector is a subsequence of U of length m.
        The list comprehension iterates over all possible starting points i in the time series to create these subsequences

        here C is A list where each element corresponds to a vector x_i from the list x
        For each vector x_i, it calculates the proportion of other vectors x_j in x that are within a distance r of x_i
        len([1 for x_j in x if _maxdist(x_i, x_j) <= r]): Counts how many vectors x_j
        have a maximum distance from x_i that is less than or equal to r.
        Dividing by (N - m + 1.0) normalizes this count by the total number of vectors.

        Returns: The mean of the logarithms of the proportions, which is the output of the _phi(m) function
        """
        x = [[U[j] for j in range(i, i + m - 1 + 1)] for i in range(N - m + 1)]
        C = [len([1 for x_j in x if _maxdist(x_i, x_j) <= r]) / (N - m + 1.0) for x_i in x]
        return (N - m + 1.0)**(-1) * sum(np.log(C))

    # N: The length of the input time series U.
    N = len(U)

    #Computes the Approximate Entropy (ApEn) by taking the absolute difference between _phi(m+1) and _phi(m).
    return abs(_phi(m+1) - _phi(m))
```

# Summary
## The ApEn function calculates the Approximate Entropy, which quantifies the amount of regularity and unpredictability in a time series. It does this by:

## Constructing vectors of length m and m+1 from the time series.
## Measuring how many pairs of these vectors remain close to each other (within a threshold r).
## Comparing the logarithmic average of these measures for vectors of length m and m+1 to derive the entropy.

```python
# A measure of complexity in the time series; a lower value indicates more regularity, while a higher value indicates more complexity.
def SampEn(U, m, r):
    """Compute Sample entropy
    U: The input time series (a list or array of data points).
    m: The embedding dimension, which defines the length of sequences to compare in the time series.
    r: The tolerance level, which defines the threshold for similarity between sequences."""
    def _maxdist(x_i, x_j):
        """_maxdist: This inner function computes the maximum distance between two vectors x_i and x_j.
        x_i: The first vector (a sequence of length m).
        x_j: The second vector (another sequence of length m).
        Returns: The maximum absolute difference between corresponding elements of x_i and x_j.
        """
        # Pairs corresponding elements of x_i and x_j using zip Computes the absolute difference between elements
        # Finds the maximum of these absole differences.
        return max([abs(ua - va) for ua, va in zip(x_i, x_j)])

    def _phi(m):
        """
        This inner function computes a value used in the calculation of Approximate Entropy.
        m: The embedding dimension.
        x: A list of vectors, where each vector is a subsequence of U of length m.
        The list comprehension iterates over all possible starting points i in the time series to create these subsequences

        C: A list where each element corresponds to a vector x_i from the list x.
        For each vector x_i, it counts how many other vectors x_j (where i != j) in x have a maximum distance from x_i that is less than 
        or equal to r.
        This count represents the number of vectors that are similar to x_i within the specified tolerance r.

        Returns: The mean of the logarithms of the proportions, which is the output of the _phi(m) function
        """
        x = [[U[j] for j in range(i, i + m - 1 + 1)] for i in range(N - m + 1)]
        C = [len([1 for j in range(len(x)) if i != j and _maxdist(x[i], x[j]) <= r]) for i in range(len(x))]

        # Returns the sum of the counts in C, representing the total number of similar vector pairs for the given embedding dimension m.
        return sum(C)


    # N: The length of the input time series U.
    N = len(U)

    # Computes the Sample Entropy (SampEn) by taking the negative natural logarithm of the ratio of _phi(m+1) to _phi(m).
    #SampEn: A measure of complexity in the time series; a lower value indicates more regularity, while a higher value indicates more complexity.
    return -np.log(_phi(m+1) / _phi(m))
```

# The SampEn function calculates the Sample Entropy, which quantifies the complexity or irregularity in a time series. It does this by:

## Constructing vectors of length m and m+1 from the time series.
## Measuring how many pairs of these vectors remain close to each other (within a threshold r), while excluding self-matches.
## Comparing the sum of these measures for vectors of length m and m+1 to derive the entropy. The function returns the negative logarithm of this ratio, which represents the Sample Entropy.
